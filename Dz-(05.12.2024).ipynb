Задание 1
Создайте программу, хранящую информацию о великих баскетболистах. Нужно хранить ФИО баскетболиста и
его рост. Требуется реализовать возможность добавления,
удаления, поиска, замены данных. Используйте словарь
для хранения информации.

```python
# Программа для управления данными о баскетболистах

def add_player(players):
    """Функция добавляет нового игрока"""
    name = input("Введите имя игрока: ")
    height = int(input(f"Введите рост {name} (в см): "))
    
    if name in players:
        print(f"Игрок с именем {name} уже существует.")
    else:
        players[name] = height
        print(f"Игрок {name} добавлен успешно!")
        
def delete_player(players):
    """Функция удаляет игрока по имени"""
    name = input("Введите имя игрока, которого хотите удалить: ")
    
    if name not in players:
        print(f"Игрока с именем {name} не найдено.")
    else:
        del players[name]
        print(f"Игрок {name} был удален.")
        
def search_player(players):
    """Функция ищет игрока по имени и выводит его рост"""
    name = input("Введите имя игрока, которого ищете: ")
    
    if name not in players:
        print(f"Игрока с именем {name} не найдено.")
    else:
        print(f"{name}: {players[name]} см")
        
def replace_player_height(players):
    """Функция заменяет рост у существующего игрока"""
    name = input("Введите имя игрока, чей рост вы хотите изменить: ")
    
    if name not in players:
        print(f"Игрока с именем {name} не найдено.")
    else:
        new_height = int(input(f"Введите новый рост для {name} (в см): "))
        players[name] = new_height
        print(f"Рост игрока {name} изменен на {new_height} см.")

def main():
    # Словарь для хранения данных об игроках
    players = {}
    
    while True:
        print("\nМеню:")
        print("1. Добавить игрока")
        print("2. Удалить игрока")
        print("3. Найти игрока")
        print("4. Изменить рост игрока")
        print("0. Выход")
        
        choice = input("Ваш выбор: ")
        
        if choice == "1":
            add_player(players)
        elif choice == "2":
            delete_player(players)
        elif choice == "3":
            search_player(players)
        elif choice == "4":
            replace_player_height(players)
        elif choice == "0":
            break
        else:
            print("Неверный ввод. Попробуйте еще раз.")
            
if __name__ == "__main__":
    main()
```

Задание 2
Создайте программу «Англо-французский словарь».
Нужно хранить слово на английском языке и его перевод
на французский. Требуется реализовать возможность добавления, удаления, поиска, замены данных. Используйте
словарь для хранения информации.

```python
def add_word(dictionary):
    """Функция добавляет новое слово в словарь."""
    english_word = input("Введите английское слово: ").lower().strip()
    french_translation = input("Введите перевод на французский язык: ").lower().strip()
    
    if english_word in dictionary:
        print(f"Слово '{english_word}' уже присутствует в словаре.")
    else:
        dictionary[english_word] = french_translation
        print(f"Слово '{english_word}' было успешно добавлено в словарь.")

def delete_word(dictionary):
    """Функция удаляет слово из словаря."""
    english_word = input("Введите английское слово, которое хотите удалить: ").lower().strip()
    
    if english_word not in dictionary:
        print(f"Слова '{english_word}' нет в словаре.")
    else:
        del dictionary[english_word]
        print(f"Слово '{english_word}' было удалено из словаря.")

def find_word(dictionary):
    """Функция находит слово в словаре и возвращает его перевод."""
    english_word = input("Введите английское слово, которое хотите найти: ").lower().strip()
    
    if english_word not in dictionary:
        print(f"Слова '{english_word}' нет в словаре.")
    else:
        translation = dictionary[english_word]
        print(f"Перевод слова '{english_word}' на французский: {translation}")

def replace_word(dictionary):
    """Функция заменяет перевод английского слова на другой."""
    english_word = input("Введите английское слово, перевод которого хотите изменить: ").lower().strip()
    
    if english_word not in dictionary:
        print(f"Слова '{english_word}' нет в словаре.")
    else:
        new_translation = input(f"Введите новый перевод для слова '{english_word}': ").lower().strip()
        dictionary[english_word] = new_translation
        print(f"Перевод слова '{english_word}' был изменён на '{new_translation}'.")

def main():
    # Словарь для хранения переводов
    dictionary = {}
    
    while True:
        print("\nМеню:")
        print("1. Добавить слово")
        print("2. Удалить слово")
        print("3. Найти слово")
        print("4. Изменить перевод")
        print("0. Выход")
        
        choice = input("Ваш выбор: ")
        
        if choice == "1":
            add_word(dictionary)
        elif choice == "2":
            delete_word(dictionary)
        elif choice == "3":
            find_word(dictionary)
        elif choice == "4":
            replace_word(dictionary)
        elif choice == "0":
            break
        else:
            print("Неверный ввод. Попробуйте ещё раз.")

if __name__ == "__main__":
    main()
```

Задание 3
Создайте программу «Фирма». Нужно хранить информацию о человеке: ФИО, телефон, рабочий email,
название должности, номер кабинета, skype. Требуется
реализовать возможность добавления, удаления, поиска, замены данных. Используйте словарь для хранения
информации.

```python
def add_employee(employees):
    """Функция добавляет нового сотрудника"""
    id_ = len(employees) + 1  # Генерируем уникальный ID
    full_name = input("Введите полное имя сотрудника: ")
    phone_number = input("Введите номер телефона сотрудника: ")
    work_email = input("Введите рабочий email сотрудника: ")
    position = input("Введите должность сотрудника: ")
    office_number = input("Введите номер кабинета сотрудника: ")
    skype = input("Введите Skype сотрудника: ")
    
    employee_data = {
        'full_name': full_name,
        'phone_number': phone_number,
        'work_email': work_email,
        'position': position,
        'office_number': office_number,
        'skype': skype
    }
    
    employees[id_] = employee_data
    print(f"Сотрудник с ID {id_} добавлен успешно!")

def delete_employee(employees):
    """Функция удаляет сотрудника по ID"""
    id_ = int(input("Введите ID сотрудника, которого хотите удалить: "))
    
    if id_ not in employees:
        print(f"Сотрудника с ID {id_} не найдено.")
    else:
        del employees[id_]
        print(f"Сотрудник с ID {id_} был удален.")

def search_employee(employees):
    """Функция ищет сотрудника по ID и выводит всю информацию о нем"""
    id_ = int(input("Введите ID сотрудника, которого ищете: "))
    
    if id_ not in employees:
        print(f"Сотрудника с ID {id_} не найдено.")
    else:
        employee_data = employees[id_]
        print(f"\nИнформация о сотруднике с ID {id_}:\n"
              f"Полное имя: {employee_data['full_name']}\n"
              f"Телефон: {employee_data['phone_number']}\n"
              f"Рабочий email: {employee_data['work_email']}\n"
              f"Должность: {employee_data['position']}\n"
              f"Номер кабинета: {employee_data['office_number']}\n"
              f"Skype: {employee_data['skype']}")

def update_employee(employees):
    """Функция обновляет информацию о сотруднике по ID"""
    id_ = int(input("Введите ID сотрудника, информацию о котором хотите обновить: "))
    
    if id_ not in employees:
        print(f"Сотрудника с ID {id_} не найдено.")
    else:
        employee_data = employees[id_]
        field_to_update = input("Что именно вы хотите обновить?\n"
                                "1. Полное имя\n"
                                "2. Телефон\n"
                                "3. Рабочий email\n"
                                "4. Должность\n"
                                "5. Номер кабинета\n"
                                "6. Skype\n"
                                "Выберите пункт: ")
        
        if field_to_update == '1':
            new_value = input("Введите новое значение для полного имени: ")
            employee_data['full_name'] = new_value
        elif field_to_update == '2':
            new_value = input("Введите новый номер телефона: ")
            employee_data['phone_number'] = new_value
        elif field_to_update == '3':
            new_value = input("Введите новый рабочий email: ")
            employee_data['work_email'] = new_value
        elif field_to_update == '4':
            new_value = input("Введите новую должность: ")
            employee_data['position'] = new_value
        elif field_to_update == '5':
            new_value = input("Введите новый номер кабинета: ")
            employee_data['office_number'] = new_value
        elif field_to_update == '6':
            new_value = input("Введите новый Skype: ")
            employee_data['skype'] = new_value
        else:
            print("Некорректный ввод. Пожалуйста, выберите корректный пункт.")
            return
        
        print(f"Информация о сотруднике с ID {id_} была успешно обновлена.")

def main():
    # Словарь для хранения всех сотрудников
    employees = {}
    
    while True:
        print("\nМеню:")
        print("1. Добавить сотрудника")
        print("2. Удалить сотрудника")
        print("3. Найти сотрудника")
        print("4. Обновить информацию о сотруднике")
        print("0. Выход")
        
        choice = input("Ваш выбор: ")
        
        if choice == "1":
            add_employee(employees)
        elif choice == "2":
            delete_employee(employees)
        elif choice == "3":
            search_employee(employees)
        elif choice == "4":
            update_employee(employees)
        elif choice == "0":
            break
        else:
            print("Неверный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

Задание 4
Создайте программу «Книжная коллекция». Нужно
хранить информацию о книгах: автор, название книги,
жанр, год выпуска, количество страниц, издательство.
Требуется реализовать возможность добавления, удаления, поиска, замены данных. Используйте словарь для
хранения информации.

```python
def add_book(books):
    """Функция добавляет новую книгу"""
    id_ = len(books) + 1  # Генерируем уникальный ID
    author = input("Введите автора книги: ")
    title = input("Введите название книги: ")
    genre = input("Введите жанр книги: ")
    year = int(input("Введите год выпуска книги: "))
    pages = int(input("Введите количество страниц: "))
    publisher = input("Введите издательство: ")
    
    book_data = {
        'author': author,
        'title': title,
        'genre': genre,
        'year': year,
        'pages': pages,
        'publisher': publisher
    }
    
    books[id_] = book_data
    print(f"Книга с ID {id_} добавлена успешно!")

def delete_book(books):
    """Функция удаляет книгу по ID"""
    id_ = int(input("Введите ID книги, которую хотите удалить: "))
    
    if id_ not in books:
        print(f"Книги с ID {id_} не найдено.")
    else:
        del books[id_]
        print(f"Книга с ID {id_} была удалена.")

def search_book(books):
    """Функция ищет книгу по ID и выводит всю информацию о ней"""
    id_ = int(input("Введите ID книги, которую ищете: "))
    
    if id_ not in books:
        print(f"Книги с ID {id_} не найдено.")
    else:
        book_data = books[id_]
        print(f"\nИнформация о книге с ID {id_}:\n"
              f"Автор: {book_data['author']}\n"
              f"Название: {book_data['title']}\n"
              f"Жанр: {book_data['genre']}\n"
              f"Год выпуска: {book_data['year']}\n"
              f"Количество страниц: {book_data['pages']}\n"
              f"Издательство: {book_data['publisher']}")

def update_book(books):
    """Функция обновляет информацию о книге по ID"""
    id_ = int(input("Введите ID книги, информацию о которой хотите обновить: "))
    
    if id_ not in books:
        print(f"Книги с ID {id_} не найдено.")
    else:
        book_data = books[id_]
        field_to_update = input("Что именно вы хотите обновить?\n"
                                "1. Автор\n"
                                "2. Название\n"
                                "3. Жанр\n"
                                "4. Год выпуска\n"
                                "5. Количество страниц\n"
                                "6. Издательство\n"
                                "Выберите пункт: ")
        
        if field_to_update == '1':
            new_value = input("Введите нового автора: ")
            book_data['author'] = new_value
        elif field_to_update == '2':
            new_value = input("Введите новое название: ")
            book_data['title'] = new_value
        elif field_to_update == '3':
            new_value = input("Введите новый жанр: ")
            book_data['genre'] = new_value
        elif field_to_update == '4':
            new_value = int(input("Введите новый год выпуска: "))
            book_data['year'] = new_value
        elif field_to_update == '5':
            new_value = int(input("Введите новое количество страниц: "))
            book_data['pages'] = new_value
        elif field_to_update == '6':
            new_value = input("Введите новое издательство: ")
            book_data['publisher'] = new_value
        else:
            print("Некорректный ввод. Пожалуйста, выберите корректный пункт.")
            return
        
        print(f"Информация о книге с ID {id_} была успешно обновлена.")

def main():
    # Словарь для хранения всех книг
    books = {}
    
    while True:
        print("\nМеню:")
        print("1. Добавить книгу")
        print("2. Удалить книгу")
        print("3. Найти книгу")
        print("4. Обновить информацию о книге")
        print("0. Выход")
        
        choice = input("Ваш выбор: ")
        
        if choice == "1":
            add_book(books)
        elif choice == "2":
            delete_book(books)
        elif choice == "3":
            search_book(books)
        elif choice == "4":
            update_book(books)
        elif choice == "0":
            break
        else:
            print("Неверный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
