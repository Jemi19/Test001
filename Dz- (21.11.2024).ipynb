Задание 1
Написать программу «справочник».Создать два списка
целых. Один список хранит идентификационные коды,
второй — телефонные номера. Реализовать меню для
пользователя:
■ Отсортировать по идентификационным кодам;
■ Отсортировать по номерам телефона;
■ Вывести список пользователей с кодами и телефонами;
■ Выход.

```python
def main():
    id_codes = [123, 456, 789, 101, 121, 213]
    phone_numbers = ["+79123456789", "+79234567891", "+79345678912", "+79456789123", "+79567891234", "+79678912345"]

    while True:
        print_menu()
        choice = input("Выберите действие: ")

        if choice == '1':
            sort_by_id(id_codes, phone_numbers)
        elif choice == '2':
            sort_by_phone(phone_numbers, id_codes)
        elif choice == '3':
            display_users(id_codes, phone_numbers)
        elif choice == 'q':
            break
        else:
            print("Неверный ввод. Попробуйте снова.")

def print_menu():
    print("\nМеню:")
    print("1. Отсортировать по идентификационным кодам")
    print("2. Отсортировать по номерам телефона")
    print("3. Вывести список пользователей с кодами и телефонами")
    print("q. Выход\n")

def sort_by_id(id_codes, phone_numbers):
    zipped = zip(id_codes, phone_numbers)
    sorted_zipped = sorted(zipped)
    id_codes, phone_numbers = zip(*sorted_zipped)
    print("Список отсортирован по идентификационным кодам.")

def sort_by_phone(phone_numbers, id_codes):
    zipped = zip(phone_numbers, id_codes)
    sorted_zipped = sorted(zipped)
    phone_numbers, id_codes = zip(*sorted_zipped)
    print("Список отсортирован по номерам телефона.")

def display_users(id_codes, phone_numbers):
    for code, number in zip(id_codes, phone_numbers):
        print(f"Код: {code}, Телефон: {number}")

if __name__ == "__main__":
    main()
```

Задание 2
Написать программу «книги». Создать два списка
годы выпуска. Реализовать меню для пользователя:
■ Отсортировать по названию книг;
■ Отсортировать по годам выпуска;
■ Вывести список книг с названиями и годами выпуска;
■ Выход;

```python
def main():
    book_titles = ["Война и мир", "Преступление и наказание", "Мастер и Маргарита", "1984", "О дивный новый мир"]
    release_years = [1869, 1866, 1967, 1949, 1932]

    while True:
        print_menu()
        choice = input("Выберите действие: ")

        if choice == '1':
            sort_by_title(book_titles, release_years)
        elif choice == '2':
            sort_by_year(release_years, book_titles)
        elif choice == '3':
            display_books(book_titles, release_years)
        elif choice == 'q':
            break
        else:
            print("Неверный ввод. Попробуйте снова.")

def print_menu():
    print("\nМеню:")
    print("1. Отсортировать по названию книг")
    print("2. Отсортировать по годам выпуска")
    print("3. Вывести список книг с названиями и годами выпуска")
    print("q. Выход\n")

def sort_by_title(titles, years):
    zipped = zip(titles, years)
    sorted_zipped = sorted(zipped)
    titles, years = zip(*sorted_zipped)
    print("Список отсортирован по названиям книг.")

def sort_by_year(years, titles):
    zipped = zip(years, titles)
    sorted_zipped = sorted(zipped)
    years, titles = zip(*sorted_zipped)
    print("Список отсортирован по годам выпуска.")

def display_books(titles, years):
    for title, year in zip(titles, years):
        print(f"Книга: {title}, Год выпуска: {year}")

if __name__ == "__main__":
    main()
```
