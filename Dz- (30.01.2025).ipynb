Задание 1
Пользователь вводит с клавиатуры набор чисел. Полученные числа необходимо сохранить в список (тип
списка нужно выбрать в зависимости от поставленной
ниже задачи). После чего нужно показать меню, в котором
предложить пользователю набор пунктов:
1. Добавить новое число в список (если такое число существует в списке, нужно вывести сообщение пользователю об этом, без добавления числа).
2. Удалить все вхождения числа из списка (пользователь
вводит с клавиатуры число для удаления)
3. Показать содержимое списка (в зависимости от выбора пользователя список нужно показать с начала
или с конца)
4. Проверить есть ли значение в списке
5. Заменить значение в списке (пользователь определяет заменить ли только первое вхождение или все
вхождения)
В зависимости от выбора пользователя выполняется
действие, после чего меню отображается снова

```python
def main():
    # Вводим данные от пользователя и сохраняем их в список
    numbers = list(map(int, input("Введите набор чисел через пробел: ").split()))
    
    while True:
        print("\nМеню:")
        print("1. Добавить новое число в список")
        print("2. Удалить все вхождения числа из списка")
        print("3. Показать содержимое списка")
        print("4. Проверить наличие значения в списке")
        print("5. Заменить значение в списке")
        print("6. Выход")
        
        choice = int(input("Выберите пункт меню: "))
        
        if choice == 1:
            new_number = int(input("Введите число для добавления: "))
            
            if new_number in numbers:
                print(f"Число {new_number} уже присутствует в списке.")
            else:
                numbers.append(new_number)
                print(f"Число {new_number} добавлено в список.")
                
        elif choice == 2:
            number_to_remove = int(input("Введите число для удаления: "))
            
            try:
                while number_to_remove in numbers:
                    numbers.remove(number_to_remove)
                print(f"Все вхождения числа {number_to_remove} удалены из списка.")
            except ValueError:
                print(f"Числа {number_to_remove} нет в списке.")
                
        elif choice == 3:
            show_from_start = input("Показать список с начала? (Y/N): ")
            
            if show_from_start.upper() == 'Y':
                for num in numbers:
                    print(num)
            else:
                for num in reversed(numbers):
                    print(num)
                    
        elif choice == 4:
            check_number = int(input("Введите число для проверки: "))
            
            if check_number in numbers:
                print(f"Число {check_number} присутствует в списке.")
            else:
                print(f"Числа {check_number} нет в списке.")
                
        elif choice == 5:
            old_value = int(input("Введите старое значение: "))
            new_value = int(input("Введите новое значение: "))
            replace_all = input("Заменить все вхождения? (Y/N): ")
            
            if replace_all.upper() == 'Y':
                while old_value in numbers:
                    index = numbers.index(old_value)
                    numbers[index] = new_value
                print(f"Все вхождения числа {old_value} заменены на {new_value}.")
            else:
                if old_value in numbers:
                    index = numbers.index(old_value)
                    numbers[index] = new_value
                    print(f"Первое вхождение числа {old_value} заменено на {new_value}.")
                else:
                    print(f"Числа {old_value} нет в списке.")
        elif choice == 6:
            break
        else:
            print("Неверный выбор. Попробуйте еще раз.")

if __name__ == "__main__":
    main()
```

Задание 2
Реализуйте класс стека для работы со строками (стек
строк).
Стек должен иметь фиксированный размер.
Реализуйте набор операций для работы со стеком:
■ помещение строки в стек;
■ выталкивание строки из стека;
■ подсчет количества строк в стеке;
■ проверку пустой ли стек;
■ проверку полный ли стек;
■ очистку стека;
■ получение значения без выталкивания верхней строки
из стека.
При старте приложения нужно отобразить меню с
помощью, которого пользователь может выбрать необходимую операцию.

```python
class StringStack:
    def __init__(self, max_size=10):
        self.max_size = max_size
        self.stack = []
        self.size = 0

    def push(self, item):
        """Помещаем строку в стек."""
        if self.is_full():
            raise IndexError("Стек полон!")
        self.stack.append(item)
        self.size += 1

    def pop(self):
        """Извлекаем строку из стека."""
        if self.is_empty():
            raise IndexError("Стек пуст!")
        self.size -= 1
        return self.stack.pop()

    def size(self):
        """Подсчитываем количество элементов в стеке."""
        return self.size

    def is_empty(self):
        """Проверяем, является ли стек пустым."""
        return self.size == 0

    def is_full(self):
        """Проверяем, является ли стек полным."""
        return self.size >= self.max_size

    def clear(self):
        """Очищаем стек."""
        self.stack.clear()
        self.size = 0

    def peek(self):
        """Получаем верхнюю строку без извлечения."""
        if self.is_empty():
            raise IndexError("Стек пуст!")
        return self.stack[-1]

# Меню для взаимодействия с пользователем
def menu(stack):
    while True:
        print("\nМеню операций со стеком:")
        print("1. Поместить строку в стек")
        print("2. Извлечь строку из стека")
        print("3. Посчитать количество строк в стеке")
        print("4. Проверить, пуст ли стек")
        print("5. Проверить, полон ли стек")
        print("6. Очистить стек")
        print("7. Получить верхнюю строку без извлечения")
        print("8. Выход")

        choice = int(input("Выберите номер операции: "))

        if choice == 1:
            string = input("Введите строку для помещения в стек: ")
            try:
                stack.push(string)
                print(f'Строка "{string}" успешно помещена в стек.')
            except IndexError as e:
                print(e)

        elif choice == 2:
            try:
                popped_string = stack.pop()
                print(f'Из стека извлечена строка: "{popped_string}".')
            except IndexError as e:
                print(e)

        elif choice == 3:
            count = stack.size()
            print(f"В стеке находится {count} строк(а/ы).")

        elif choice == 4:
            if stack.is_empty():
                print("Стек пуст.")
            else:
                print("Стек не пуст.")

        elif choice == 5:
            if stack.is_full():
                print("Стек полон.")
            else:
                print("Стек не полон.")

        elif choice == 6:
            stack.clear()
            print("Стек очищен.")

        elif choice == 7:
            try:
                top_string = stack.peek()
                print(f'Верхняя строка в стеке: "{top_string}".')
            except IndexError as e:
                print(e)

        elif choice == 8:
            break

        else:
            print("Неправильный выбор. Попробуйте ещё раз.")

# Запуск программы
if __name__ == "__main__":
    my_stack = StringStack(max_size=5)  # Создаем стек размером 5
    menu(my_stack)
```
