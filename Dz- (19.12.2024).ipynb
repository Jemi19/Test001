Задание №1 Class “Книга”
Создайте класс Книга с атрибутами:
название (строка),
автор (строка),
год_издания (целое число),
количество_страниц (целое число),
цена (вещественное число),
Добавьте в класс Книга метод рассчитать_стоимость,
который будет возвращать стоимость книги, исходя из
ее цены и количества страниц. Формула для расчета
стоимости: цена * количество_страниц / 100.
Добавьте в класс Книга метод
изменить_количество_страниц, который будет
принимать новое количество страниц и изменять
соответствующий атрибут.

```python
class Book:
    def __init__(self, title, author, year_of_publication, number_of_pages, price):
        self.title = title
        self.author = author
        self.year_of_publication = year_of_publication
        self.number_of_pages = number_of_pages
        self.price = price

    def calculate_cost(self):
        """Метод для расчёта стоимости книги."""
        cost = self.price * self.number_of_pages / 100
        return round(cost, 2)

    def change_number_of_pages(self, new_number_of_pages):
        """Метод для изменения количества страниц."""
        self.number_of_pages = new_number_of_pages

# Пример использования класса
book = Book('Война и мир', 'Лев Толстой', 1869, 1225, 500)
print(f'Название книги: {book.title}')
print(f'Автор: {book.author}')
print(f'Год издания: {book.year_of_publication}')
print(f'Количество страниц: {book.number_of_pages}')
print(f'Цена за страницу: {book.price:.2f}')
print(f'Общая стоимость книги: {book.calculate_cost():.2f}')

# Изменяем количество страниц
new_pages = 1500
book.change_number_of_pages(new_pages)
print(f'\nИзмененное количество страниц: {book.number_of_pages}')
print(f'Новая общая стоимость книги: {book.calculate_cost():.2f}')
```

Задание №2 Class “Counter”
Создайте класс Счетчик, который будет моделировать
счетчик, используемый для подсчета чего-либо. Класс
должен иметь следующие свойства и методы:
Приватное свойство _значение, которое хранит
текущее значение счетчика.
Метод увеличить(), который увеличивает значение
счетчика на 1.
Метод уменьшить(), который уменьшает значение
счетчика на 1.
Метод получить_значение(), который возвращает
текущее значение счетчика.
Метод сбросить(), который сбрасывает значение
счетчика на 0.
Дополнительные требования:
Значение счетчика не может быть меньше 0.
При попытке уменьшить значение счетчика ниже 0,
оно должно оставаться равным 0.

```python
class Counter:
    def __init__(self):
        # Приватное свойство для хранения текущего значения счетчика
        self._value = 0

    def increase(self):
        """Увеличиваем значение счетчика на 1"""
        self._value += 1

    def decrease(self):
        """Уменьшаем значение счетчика на 1, но не ниже нуля"""
        if self._value > 0:
            self._value -= 1

    def get_value(self):
        """Возвращаем текущее значение счетчика"""
        return self._value

    def reset(self):
        """Сбрасываем значение счетчика на 0"""
        self._value = 0

# Пример использования класса
counter = Counter()

# Увеличение значения счетчика
for _ in range(10):
    counter.increase()

# Получение текущего значения
current_value = counter.get_value()
print(f"Текущее значение счетчика: {current_value}")

# Уменьшение значения счетчика
for _ in range(15):
    counter.decrease()

# Получение текущего значения
current_value = counter.get_value()
print(f"Текущее значение счетчика: {current_value}")

# Сброс значения счетчика
counter.reset()
current_value = counter.get_value()
print(f"Текущее значение счетчика: {current_value}")
```

Задание №3 Class “BankAcc”
Создайте класс БанковскийСчет, который будет
моделировать банковский счет. Класс должен иметь
следующие свойства и методы:
Приватное свойство _баланс, которое хранит текущий
баланс счета.
Метод положить_деньги(сумма), который увеличивает
баланс на указанную сумму.
Метод снять_деньги(сумма), который уменьшает
баланс на указанную сумму. Если на счете
недостаточно средств, должно выводиться сообщение
об ошибке.
Метод получить_баланс(), который возвращает
текущий баланс счета.
Метод перевести_деньги(другой_счет, сумма),
который переводит указанную сумму с текущего счета
на другой счет. Если на текущем счете недостаточно
средств, должно выводиться сообщение об ошибке.
Дополнительные требования:
Баланс не может быть отрицательным.
Сумма, передаваемая в методы, должна быть
положительной.

```python
class BankAccount:
    def __init__(self):
        # Приватное свойство для хранения баланса
        self._balance = 0

    def deposit_money(self, amount):
        """Кладём деньги на счёт."""
        if amount <= 0:
            print("Сумма должна быть положительной.")
            return
        self._balance += amount
        print(f"Успешно пополнено: {amount}. Текущий баланс: {self._balance}")

    def withdraw_money(self, amount):
        """Снимаем деньги со счёта."""
        if amount <= 0:
            print("Сумма должна быть положительной.")
            return
        if amount > self._balance:
            print("Недостаточно средств на счету.")
            return
        self._balance -= amount
        print(f"Снятие средств выполнено: {amount}. Остаток на счету: {self._balance}")

    def get_balance(self):
        """Получаем текущий баланс."""
        return self._balance

    def transfer_money(self, other_account, amount):
        """Перевод денег на другой счёт."""
        if amount <= 0:
            print("Сумма должна быть положительной.")
            return
        if amount > self._balance:
            print("Недостаточно средств на счету.")
            return
        self.withdraw_money(amount)
        other_account.deposit_money(amount)
        print(f"Перечислено {amount} на другой счёт. Ваш остаток: {self._balance}")

# Пример использования класса
account1 = BankAccount()
account2 = BankAccount()

# Пополнение первого счёта
account1.deposit_money(1000)

# Перевод части суммы на второй счёт
account1.transfer_money(account2, 500)

# Проверка балансов обоих счетов
print(f"Баланс первого счёта: {account1.get_balance()}")
print(f"Баланс второго счёта: {account2.get_balance()}")

# Попытка снятия большей суммы, чем есть на счету
account1.withdraw_money(600)
```
