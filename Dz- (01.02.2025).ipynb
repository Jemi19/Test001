Задание 1
Есть некоторый словарь, который хранит названия
стран и столиц. Название страны используется в качестве
ключа, название столицыв качестве значения. Необходимо
реализовать: добавление данных, удаление данных, поиск
данных, редактирование данных, сохранение и загрузку
данных (используя упаковку и распаковку).

```python
import pickle

# Инициализация словаря
countries_capitals = {}

def add_data():
    country = input("Введите страну: ")
    capital = input("Введите столицу: ")
    countries_capitals[country] = capital
    print(f"Добавлена запись: страна '{country}' со столицей '{capital}'.")

def delete_data():
    country = input("Введите страну для удаления: ")
    if country in countries_capitals:
        del countries_capitals[country]
        print(f"Страна '{country}' удалена.")
    else:
        print(f"Страна '{country}' отсутствует в базе данных.")

def search_data():
    country = input("Введите страну для поиска: ")
    if country in countries_capitals:
        capital = countries_capitals[country]
        print(f"Столицей страны '{country}' является '{capital}'.")
    else:
        print(f"Страна '{country}' отсутствует в базе данных.")

def edit_data():
    country = input("Введите страну для редактирования: ")
    if country in countries_capitals:
        new_capital = input("Введите новую столицу: ")
        countries_capitals[country] = new_capital
        print(f"Столица страны '{country}' изменена на '{new_capital}'.")
    else:
        print(f"Страна '{country}' отсутствует в базе данных.")

def save_data():
    with open('data.pkl', 'wb') as file:
        pickle.dump(countries_capitals, file)
    print("Данные успешно сохранены.")

def load_data():
    global countries_capitals
    try:
        with open('data.pkl', 'rb') as file:
            countries_capitals = pickle.load(file)
        print("Данные успешно загружены.")
    except FileNotFoundError:
        print("Файл с данными не найден. Начните с добавления новых записей.")

def display_menu():
    print("\nМеню:")
    print("1. Добавить данные")
    print("2. Удалить данные")
    print("3. Найти данные")
    print("4. Редактировать данные")
    print("5. Сохранить данные")
    print("6. Загрузить данные")
    print("7. Выход")

def main():
    while True:
        display_menu()
        choice = int(input("Выберите пункт меню: "))

        if choice == 1:
            add_data()
        elif choice == 2:
            delete_data()
        elif choice == 3:
            search_data()
        elif choice == 4:
            edit_data()
        elif choice == 5:
            save_data()
        elif choice == 6:
            load_data()
        elif choice == 7:
            break
        
        else:
            print("Неверный выбор. Попробуйте еще раз.")

if __name__ == "__main__":
    main()
```

Задание 2
Есть некоторый словарь, который хранит названия
музыкальных групп(исполнителей) и альбомов. Название группы используется в качестве ключа, название
альбомов в качестве значения. Необходимо реализовать:
добавление данных, удаление данных, поиск данных,
редактирование данных, сохранение и загрузку данных
(используя упаковку и распаковку).

```python
import pickle

# Инициализация словаря
bands_albums = {}

def add_data():
    band = input("Введите название музыкальной группы: ")
    album = input("Введите название альбома: ")
    bands_albums.setdefault(band, []).append(album)
    print(f"Альбом '{album}' добавлен для группы '{band}'.")

def delete_data():
    band = input("Введите название группы для удаления: ")
    if band in bands_albums:
        albums = bands_albums[band]
        if len(albums) > 0:
            album = input(f"Выберите альбом для удаления ({', '.join(albums)}): ")
            if album in albums:
                albums.remove(album)
                print(f"Альбом '{album}' удален для группы '{band}'.")
            else:
                print(f"Альбом '{album}' не найден для группы '{band}'.")
        else:
            print(f"У группы '{band}' нет альбомов.")
    else:
        print(f"Группа '{band}' отсутствует в базе данных.")

def search_data():
    band = input("Введите название группы для поиска: ")
    if band in bands_albums:
        albums = bands_albums[band]
        if len(albums) > 0:
            print(f"Альбомы группы '{band}': {', '.join(albums)}.")
        else:
            print(f"У группы '{band}' нет альбомов.")
    else:
        print(f"Группа '{band}' отсутствует в базе данных.")

def edit_data():
    band = input("Введите название группы для редактирования: ")
    if band in bands_albums:
        albums = bands_albums[band]
        if len(albums) > 0:
            album = input(f"Выберите альбом для редактирования ({', '.join(albums)}): ")
            if album in albums:
                new_album = input("Введите новое название альбома: ")
                index = albums.index(album)
                albums[index] = new_album
                print(f"Альбом '{album}' изменен на '{new_album}'.")
            else:
                print(f"Альбом '{album}' не найден для группы '{band}'.")
        else:
            print(f"У группы '{band}' нет альбомов.")
    else:
        print(f"Группа '{band}' отсутствует в базе данных.")

def save_data():
    with open('data.pkl', 'wb') as file:
        pickle.dump(bands_albums, file)
    print("Данные успешно сохранены.")

def load_data():
    global bands_albums
    try:
        with open('data.pkl', 'rb') as file:
            bands_albums = pickle.load(file)
        print("Данные успешно загружены.")
    except FileNotFoundError:
        print("Файл с данными не найден. Начните с добавления новых записей.")

def display_menu():
    print("\nМеню:")
    print("1. Добавить данные")
    print("2. Удалить данные")
    print("3. Найти данные")
    print("4. Редактировать данные")
    print("5. Сохранить данные")
    print("6. Загрузить данные")
    print("7. Выход")

def main():
    while True:
        display_menu()
        choice = int(input("Выберите пункт меню: "))

        if choice == 1:
            add_data()
        elif choice == 2:
            delete_data()
        elif choice == 3:
            search_data()
        elif choice == 4:
            edit_data()
        elif choice == 5:
            save_data()
        elif choice == 6:
            load_data()
        elif choice == 7:
            break
        else:
            print("Неверный выбор. Попробуйте еще раз.")

if __name__ == "__main__":
    main()
```
