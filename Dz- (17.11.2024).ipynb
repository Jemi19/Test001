Задание 1
Необходимо отсортировать первые две трети списка
в порядке возрастания, если среднее арифметическое
всех элементов больше нуля; иначе — лишь первую треть.
Остальную часть списка не сортировать, а расположить
в обратном порядке.

```python
def custom_sort(lst):
    n = len(lst)
    mean = sum(lst) / n
    
    one_third = n // 3
    two_thirds = 2 * one_third
    
    if mean > 0:
        sorted_part = sorted(lst[:two_thirds])
        unsorted_part = lst[two_thirds:]
    else:
        sorted_part = sorted(lst[:one_third])
        unsorted_part = lst[one_third:]
    
    reversed_unsorted_part = list(reversed(unsorted_part))
    
    return sorted_part + reversed_unsorted_part

# Пример использования
lst = [9, 6, 12, 15, 20, 25, 30, 35, 40, 45, 50]
result = custom_sort(lst)
print(result)
```

```python
# 1 Вариант
def main():
    grades = []

    # Запрашиваем оценки у пользователя
    for i in range(10):
        grade = int(input(f"Введите оценку №{i+1}: "))
        while not (1 <= grade <= 12):
            print("Оценка должна быть в диапазоне от 1 до 12!")
            grade = int(input(f"Введите оценку №{i+1}: "))
        grades.append(grade)

    # Основной цикл программы
    while True:
        print_menu()
        choice = input("Выберите пункт меню: ")

        if choice == '1':
            show_grades(grades)
        elif choice == '2':
            retake_exam(grades)
        elif choice == '3':
            check_scholarship(grades)
        elif choice == '4':
            sort_grades(grades)
        elif choice == 'q':
            break
        else:
            print("Неверный ввод. Попробуйте снова.")

def print_menu():
    print("\nМеню:")
    print("1. Вывод оценок")
    print("2. Пересдача экзамена")
    print("3. Выходит ли стипендия")
    print("4. Сортировка оценок")
    print("q. Выход\n")

def show_grades(grades):
    print("\nВаши оценки:")
    for i, grade in enumerate(grades, start=1):
        print(f"{i}. {grade}")

def retake_exam(grades):
    index = int(input("Введите номер оценки, которую хотите пересдать: "))
    while not (1 <= index <= len(grades)):
        print("Номер оценки должен быть в пределах списка!")
        index = int(input("Введите номер оценки, которую хотите пересдать: "))

    new_grade = int(input("Введите новую оценку: "))
    while not (1 <= new_grade <= 12):
        print("Новая оценка должна быть в диапазоне от 1 до 12!")
        new_grade = int(input("Введите новую оценку: "))

    grades[index-1] = new_grade
    print(f"Оценка под номером {index} была изменена на {new_grade}.")

def check_scholarship(grades):
    average_grade = sum(grades) / len(grades)
    if average_grade >= 10.7:
        print("Да, студент получает стипендию.")
    else:
        print("Нет, средний балл слишком низкий для получения стипендии.")

def sort_grades(grades):
    order = input("Как сортировать оценки? (a - по возрастанию, d - по убыванию): ").lower()
    if order == 'a':
        sorted_grades = sorted(grades)
    elif order == 'd':
        sorted_grades = sorted(grades, reverse=True)
    else:
        print("Неправильный выбор. Повторите попытку.")
        return

    print("\nОтсортированные оценки:")
    for i, grade in enumerate(sorted_grades, start=1):
        print(f"{i}. {grade}")

if __name__ == "__main__":
    main()
```

Задание 3
Написать программу, реализующую сортировку списка
методом усовершенствованной сортировки пузырьковым
методом. Усовершенствование состоит в том, чтобы анализировать количество перестановок на каждом шагу, если
это количество равно нулю, то продолжать сортировку
нет смысла — список отсортирован

```python
def bubble_sort_optimized(arr):
    n = len(arr)
    swapped = True  # Флаг для определения необходимости продолжения сортировки
    
    while swapped:
        swapped = False  # Предполагаем, что на этом проходе перестановок не будет
        for i in range(1, n):
            if arr[i - 1] > arr[i]:
                # Меняем элементы местами
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                swapped = True  # Перестановка произошла, продолжаем сортировку
                
        # Уменьшаем размер проверяемой области, т.к. последний элемент всегда максимальный
        n -= 1
    
    return arr

# Пример использования
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = bubble_sort_optimized(arr)
print("Отсортированный массив:", sorted_arr)
```
