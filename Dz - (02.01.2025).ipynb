Задание 1 Создайте класс Device, который содержит информацию об устройстве. 
Спомощью механизма наследования, реализуйте класс Cooffe machine 
(содержит информацию о комемашине) 
класс Blender (содержит информацию о блендере), 
класс MeatGrinder (содержит информацию о мясорубке). 
Каждый из классов должен содержать необходимые для работы методы.

```python
class Device:
    def __init__(self, brand, model, power):
        self.brand = brand
        self.model = model
        self.power = power  # Мощность в ваттах
    
    def turn_on(self):
        print(f"{self.brand} {self.model} turned on")
    
    def turn_off(self):
        print(f"{self.brand} {self.model} turned off")

    def get_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Power: {self.power}W")

class CoffeeMachine(Device):
    def __init__(self, brand, model, power, water_tank_capacity, bean_container_capacity):
        super().__init__(brand, model, power)
        self.water_tank_capacity = water_tank_capacity  # Объем водяного бака в литрах
        self.bean_container_capacity = bean_container_capacity  # Объем контейнера для зерен в граммах
        
    def brew_coffee(self, amount):
        if amount <= self.water_tank_capacity and amount <= self.bean_container_capacity:
            print(f"Brewing {amount}ml of coffee...")
            self.water_tank_capacity -= amount
            self.bean_container_capacity -= amount
        else:
            print("Not enough water or beans to brew coffee!")
            
    def refill_water_tank(self, amount):
        if self.water_tank_capacity + amount <= 1000:
            self.water_tank_capacity += amount
            print(f"Refilled water tank with {amount}ml. Current capacity: {self.water_tank_capacity}ml")
        else:
            print("Water tank overflow! Cannot add more water.")

    def refill_bean_container(self, amount):
        if self.bean_container_capacity + amount <= 200:
            self.bean_container_capacity += amount
            print(f"Refilled bean container with {amount}g. Current capacity: {self.bean_container_capacity}g")
        else:
            print("Bean container overflow! Cannot add more beans.")

class Blender(Device):
    def __init__(self, brand, model, power, bowl_capacity):
        super().__init__(brand, model, power)
        self.bowl_capacity = bowl_capacity  # Объем чаши в миллилитрах
        self.contents = []  # Список ингредиентов в чаше
        
    def add_ingredient(self, ingredient):
        if len(self.contents) < self.bowl_capacity:
            self.contents.append(ingredient)
            print(f"Added {ingredient} to the blender.")
        else:
            print("Blender is full! Cannot add more ingredients.")

    def blend(self):
        if self.contents:
            print("Blending ingredients...")
            print(", ".join(self.contents))
            self.contents.clear()
        else:
            print("Nothing to blend! Please add some ingredients first.")

class MeatGrinder(Device):
    def __init__(self, brand, model, power, tray_capacity):
        super().__init__(brand, model, power)
        self.tray_capacity = tray_capacity  # Объем лотка в килограммах
        self.meat_in_tray = 0  # Текущий вес мяса в лотке
        
    def add_meat(self, weight):
        if self.meat_in_tray + weight <= self.tray_capacity:
            self.meat_in_tray += weight
            print(f"Added {weight}kg of meat. Current weight in tray: {self.meat_in_tray}kg")
        else:
            print("Tray is full! Cannot add more meat.")

    def grind(self):
        if self.meat_in_tray > 0:
            print(f"Grinding {self.meat_in_tray}kg of meat...")
            self.meat_in_tray = 0
        else:
            print("No meat to grind! Please add some meat first.")

coffee_machine = CoffeeMachine("DeLonghi", "Magnifica", 1500, 1800, 300)
coffee_machine.turn_on()
coffee_machine.get_info()
coffee_machine.brew_coffee(250)
coffee_machine.refill_water_tank(500)
coffee_machine.turn_off()

blender = Blender("Philips", "HR3655/00", 1200, 600)
blender.turn_on()
blender.add_ingredient("Banana")
blender.add_ingredient("Strawberry")
blender.blend()
blender.turn_off()

meat_grinder = MeatGrinder("Bosch", "MFW68660", 800, 5)
meat_grinder.turn_on()
meat_grinder.add_meat(2.5)
meat_grinder.grind()
meat_grinder.turn_off()

```

Задание 2
Создайте класс Ship, который содержит информацию
о корабле.
С помощью механизма наследования, реализуйте
класс Frigate (содержит информацию о фрегате), класс
Destroyer (содержит информацию об эсминце), класс
Cruiser (содержит информацию о крейсере).
Каждый из классов должен содержать необходимые
для работы методы.

```python
class Ship:
    def __init__(self, name, displacement, speed):
        self.name = name
        self.displacement = displacement  # Водоизмещение в тоннах
        self.speed = speed  # Скорость в узлах
    
    def move(self):
        print(f"{self.name} moving at {self.speed} knots")
    
    def stop(self):
        print(f"{self.name} stopped")

    def get_info(self):
        print(f"Name: {self.name}")
        print(f"Displacement: {self.displacement} tons")
        print(f"Speed: {self.speed} knots")

class Frigate(Ship):
    def __init__(self, name, displacement, speed, missile_count):
        super().__init__(name, displacement, speed)
        self.missile_count = missile_count  # Количество ракет на борту
        
    def fire_missiles(self, target):
        if self.missile_count > 0:
            print(f"Firing missiles at {target}...")
            self.missile_count -= 1
        else:
            print("Out of missiles! Reload required.")

class Destroyer(Ship):
    def __init__(self, name, displacement, speed, torpedo_count):
        super().__init__(name, displacement, speed)
        self.torpedo_count = torpedo_count  # Количество торпед на борту
        
    def launch_torpedoes(self, target):
        if self.torpedo_count > 0:
            print(f"Launching torpedoes at {target}...")
            self.torpedo_count -= 1
        else:
            print("Out of torpedoes! Reload required.")

class Cruiser(Ship):
    def __init__(self, name, displacement, speed, artillery_count):
        super().__init__(name, displacement, speed)
        self.artillery_count = artillery_count  # Количество артиллерийских установок
        
    def fire_artillery(self, target):
        if self.artillery_count > 0:
            print(f"Shooting artillery at {target}...")
        else:
            print("Artillery malfunction! Repair needed.")

# Создание объектов и их использование
frigate = Frigate("HMS Argyll", 4500, 28, 8)
frigate.move()
frigate.fire_missiles("Enemy ship")
frigate.stop()

destroyer = Destroyer("USS Zumwalt", 14500, 30, 12)
destroyer.move()
destroyer.launch_torpedoes("Submarine")
destroyer.stop()

cruiser = Cruiser("Ticonderoga", 9600, 32, 2)
cruiser.move()
cruiser.fire_artillery("Land target")
cruiser.stop()
```

Задание 3
Запрограммируйте класс Money (объект класса оперирует одной валютой) для работы с деньгами.
В классе должны быть предусмотрены поле для хранения целой части денег (доллары, евро, гривны и т.д.) и
поле для хранения копеек (центы, евроценты, копейки
и т.д.).
Реализовать методы для вывода суммы на экран, задания значений для частей. 

```python
class Money:
    def __init__(self, dollars=0, cents=0):
        self.dollars = dollars
        self.cents = cents
        self._normalize()

    def _normalize(self):
        """ Приводит сумму к нормальному виду (менее 100 центов) """
        self.dollars += self.cents // 100
        self.cents %= 100

    def set_value(self, dollars, cents):
        """ Задает значения для частей """
        self.dollars = dollars
        self.cents = cents
        self._normalize()

    def display(self):
        """ Выводит сумму на экран """
        print(f"{self.dollars}.{self.cents:02d}")

    def __str__(self):
        """ Строковое представление объекта """
        return f"{self.dollars}.{self.cents:02d}"

    def __repr__(self):
        """ Представление объекта для разработчиков """
        return f"Money({self.dollars}, {self.cents})"

    def __add__(self, other):
        """ Операция сложения """
        new_dollars = self.dollars + other.dollars
        new_cents = self.cents + other.cents
        return Money(new_dollars, new_cents)

    def __sub__(self, other):
        """ Операция вычитания """
        new_dollars = self.dollars - other.dollars
        new_cents = self.cents - other.cents
        return Money(new_dollars, new_cents)

    def __mul__(self, value):
        """ Умножение на целое число """
        total_cents = self.dollars * 100 + self.cents
        new_total_cents = total_cents * value
        new_dollars = new_total_cents // 100
        new_cents = new_total_cents % 100
        return Money(new_dollars, new_cents)

    def __truediv__(self, value):
        """ Деление на целое число """
        total_cents = self.dollars * 100 + self.cents
        new_total_cents = total_cents // value
        new_dollars = new_total_cents // 100
        new_cents = new_total_cents % 100
        return Money(new_dollars, new_cents)

money1 = Money(10, 50)
money2 = Money(20, 75)

# Вывод суммы на экран
money1.display()  # Output: 10.50
money2.display()  # Output: 20.75

# Сложение
total = money1 + money2
total.display()  # Output: 31.25

# Вычитание
difference = money2 - money1
difference.display()  # Output: 10.25

# Умножение
multiplied = money1 * 2
multiplied.display()  # Output: 21.00

# Деление
divided = money2 / 2
divided.display()  # Output: 10.37
```
